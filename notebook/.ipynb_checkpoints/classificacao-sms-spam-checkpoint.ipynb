```python
import pandas as pd 
import string
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
def preprocess_text(text):
    # 1. Min√∫sculas
    text = text.lower()

    # 2. Remover pontua√ß√£o
    text = text.translate(str.maketrans('', '', string.punctuation))

    # 3. Remover stopwords
    stop_words = set(stopwords.words('english'))
    words = text.split()
    filtered_words = [word for word in words if word not in stop_words]
    
    return " ".join(filtered_words)
```

```python
df = pd.read_csv('../data/spam.csv', encoding='latin=1')
```

```python
df.dropna(axis=1, how='any', inplace=True)
```

```python
df.columns = ['label', 'message']
```

#### Pr√©-processamento de texto

```python
df['clean_message'] = df['message'].apply(preprocess_text)
df[['message', 'clean_message']].head()
```

#### Vetoriza√ß√£o

```python
# Inicializando o vetor de contagem
vectorizer = CountVectorizer()
```

```python
# Transformando o texto
X = vectorizer.fit_transform(df['clean_message'])
```

```python
# Exibindo a forma da matriz resultante (documentos x caracter√≠sticas)
X.shape
```

```python
vocab = vectorizer.vocabulary_
#print(vocab)
```

#### Divis√£o em treino e teste

```python
# Dividindo o dataset em treino (80%) e teste (20%)
X_train, X_test, y_train, y_test = train_test_split(X, df['label'], test_size=0.2, random_state=42)
```

```python
# Verificando as formas dos conjuntos
print(X_train.shape, X_test.shape)
```

#### Treinando o modelo

```python
# Criando o modelo de Naive Bayes
model = MultinomialNB()
```

```python
# Treinando o modelo
model.fit(X_train, y_train)
```

#### Avaliando o modelo

```python
# Prevendo as classes no conjunto de teste
y_pred = model.predict(X_test)
```

```python
# Calculando a acur√°cia
print("Acur√°cia:", accuracy_score(y_test, y_pred))
```

```python
# Exibindo um relat√≥rio completo de m√©tricas
print(classification_report(y_test, y_pred))
```
